```python
# Count frequency of each ingredient
ingredient_counts = df['ingredients__(spices/veggies)'].value_counts().reset_index()
ingredient_counts.columns = ['ingredients__(spices/veggies)', 'Frequency']

# Get top 4 ingredients
top_4_ingredients = ingredient_counts.head(4)
print(top_4_ingredients)

# Plot bar chart
plt.figure(figsize=(8, 5))
sns.barplot(data=top_4_ingredients, x='ingredients__(spices/veggies)', y='Frequency', palette='cubehelix')
plt.title('Top 4 Most Common Ingredients')
plt.xlabel('ingredients__(spices/veggies)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Define spice list (customize as needed)
spice_list = ['turmeric', 'chili', 'cumin', 'coriander', 'mustard', 'cardamom', 'clove', 'fenugreek', 'asafoetida', 'cinnamon', 'pepper', 'nutmeg', 'bay leaf']

# Clean data
df = df.dropna(subset=['ingredients__(spices/veggies)'])
df['ingredients__(spices/veggies)'] = df['ingredients__(spices/veggies)'].str.strip().str.lower()

# Optional: Standardize Thali/Region info
df['thali_name_'] = df['thali_name_'].str.strip()
df['Region'] = df['thali_name_'].str.extract(r'(^[A-Za-z]+(?:[_\s][A-Za-z]+)?)')[0].str.replace('_', ' ').str.title()

# Add is_spice column
df['is_spice'] = df['ingredients__(spices/veggies)'].apply(lambda x: 1 if any(spice in x for spice in spice_list) else 0)

# Count analysis
count_summary = df['is_spice'].value_counts().reset_index()
count_summary.columns = ['Is_Spice', 'Count']
count_summary['Label'] = count_summary['Is_Spice'].map({1: 'Spice', 0: 'Non-Spice'})
print("\n🔢 Count Summary:\n", count_summary)

# Weight analysis (assuming 'Weight' column exists)
if 'Weight' in df.columns:
    weight_summary = df.groupby('is_spice')['Weight'].sum().reset_index()
    weight_summary.columns = ['Is_Spice', 'Total_Weight']
    weight_summary['Label'] = weight_summary['Is_Spice'].map({1: 'Spice', 0: 'Non-Spice'})
    print("\n⚖️ Weight Summary:\n", weight_summary)
else:
    print("\n⚠️ No 'Weight' column found — skipping weight analysis.")

# Spice/Non-Spice by Region (count)
region_count = df.groupby(['Region', 'is_spice']).size().reset_index(name='Count')
region_count['Label'] = region_count['is_spice'].map({1: 'Spice', 0: 'Non-Spice'})

# --- Visualization ---

# Count bar chart
plt.figure(figsize=(6, 4))
sns.barplot(data=count_summary, x='Label', y='Count', palette='Set2')
plt.title('Spice vs Non-Spice (Count)')
plt.tight_layout()
plt.show()

# Weight bar chart
if 'Weight' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.barplot(data=weight_summary, x='Label', y='Total_Weight', palette='Set3')
    plt.title('Spice vs Non-Spice (Total Weight)')
    plt.tight_layout()
    plt.show()

# Per Region stacked bar chart
pivot_region = region_count.pivot(index='Region', columns='Label', values='Count').fillna(0)
pivot_region.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Paired')
plt.title('Spice vs Non-Spice Count by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
